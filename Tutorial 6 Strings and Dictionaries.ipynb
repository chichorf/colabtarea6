## Sintaxis de cadenas

Las cadenas en Python se pueden definir usando comillas simples o dobles. Son funcionalmente equivalentes.

```python
x = 'Plut√≥n es un planeta'
y = "Plut√≥n es un planeta"
x == y
```

Las comillas dobles son convenientes si la cadena contiene un car√°cter de comilla simple (por ejemplo, si representa un ap√≥strofo).

Del mismo modo, es f√°cil crear una cadena que contenga comillas dobles si la envolvemos entre comillas simples:

```python
print("¬°Plut√≥n es un planeta!")
print('Mi perro se llama "Pluto')
```

Si intentamos poner un car√°cter de comilla simple dentro de una cadena de comillas simples, Python se confunde:

```python
'Pluton's a planet!'
```

Podemos arreglar esto "escapando" de la comilla simple con una barra invertida.

```python
'Pluton\'s a planet!'
```

La siguiente tabla resume algunos usos importantes del car√°cter de barra invertida.

| Lo que escribes... | Lo que obtienes | Ejemplo           | `print(ejemplo)`            |
|------------------|--------------|-----------------------|------------------------------|
| `\'`             | `'`          | `'What\'s up?'`       | `What's up?`                 |  
| `\"`             | `"`          | `"That's \"cool\""`   | `That's "cool"`              |  
| `\\`             | `\`          |  `"Mira, una monta√±a: /\\"` |  `Mira, una monta√±a: /\` |
| `\n`             |   <br/>      |   `"1\n2 3"`          |   `1`<br/>`2 3`              |

La √∫ltima secuencia, `\n`, representa el *car√°cter de nueva l√≠nea*. Hace que Python comience una nueva l√≠nea.

```python
hello = "hola\nmundo"
print(hello)
```

Adem√°s, la sintaxis de comillas triples de Python para cadenas nos permite incluir saltos de l√≠nea literalmente (es decir, simplemente presionando 'Enter' en nuestro teclado, en lugar de usar la secuencia especial '\n'). Ya hemos visto esto en las cadenas de documentaci√≥n que usamos para documentar nuestras funciones, pero podemos usarlas en cualquier lugar que queramos definir una cadena.

```python
triplequoted_hello = """hola mundo"""
print(triplequoted_hello)
triplequoted_hello == hello
```

La funci√≥n `print()` agrega autom√°ticamente un car√°cter de nueva l√≠nea a menos que especifiquemos un valor para el argumento de la palabra clave `end` que no sea el valor predeterminado de `'\n'`:

```python
print("hola")
print("mundo")
print("hola", end='')
print("pluto", end='')
```

## Las cadenas son secuencias

Las cadenas se pueden considerar como secuencias de caracteres. Casi todo lo que hemos visto que podemos hacer con una lista, tambi√©n podemos hacerlo con una cadena.

```python
# Indexaci√≥n
planet = 'Pluton'
planet[0]
```

```python
# Rebanar
planet[-3:]
```

```python
# ¬øCu√°nto mide esta cadena?
len(planet)
```

```python
# S√≠, incluso podemos recorrerlas.
[char+'! ' for char in planet]
```

Pero una forma importante en la que se diferencian de las listas es que son *inmutables*. No podemos modificarlos.

```python
planet[0] = 'B'
# planet.append tampoco funciona
```

### Ir entre cadenas y listas: `.split()` y `.join()`

`str.split()` convierte una cadena en una lista de cadenas m√°s peque√±as, interrumpiendo los espacios en blanco de forma predeterminada. Esto es √∫til para convertir una gran cadena en una lista de palabras.

```python
words = claim.split()
words
```

¬øQu√© pasa si queremos dividirnos en algo que no sea un espacio en blanco?

```python
datestr = '1956-01-31'
a√±o, mes, dia = datestr.split('-')
```

`str.join()` nos lleva en la otra direcci√≥n, cosiendo una lista de cadenas en una cadena larga, usando la cadena a la que se llam√≥ como separador.

```python
'/'.join([mes, dia, a√±o])
```

```python
# S√≠, podemos poner caracteres Unicode directamente en nuestras cadenas literales :)
' üëè '.join([word.upper() for word in words])
```

### Building strings with `.format()`

Python lets us concatenate strings with the `+` operator.

```python
planet + ', te extra√±o.'
```

*Si* queremos agregar cualquier objeto que no sea una cadena, debemos tener cuidado de llamar a `str()` en ellos primero

```python
position = 9
planet + "Siempre seras el " + position + "el planeta para m√≠."
```

```python
planet + "Siempre seras el " + str(position) + "el planeta para m√≠."
```

Esto se est√° volviendo dif√≠cil de leer y molesto de escribir. `str.format()` al rescate.

```python
"{}, siempre ser√°s el {} en√©simo planeta para m√≠.".format(planet, position)
```

¬°Mucho m√°s limpio! Llamamos a `.format()` en una "cadena de formato", donde los valores de Python que queremos insertar se representan con marcadores de posici√≥n `{}`.

Observe c√≥mo ni siquiera tuvimos que llamar a `str()` para convertir `posici√≥n` de un int. `format()` se encarga de eso por nosotros.

Si eso fuera todo lo que `format()` hiciera, a√∫n ser√≠a incre√≠blemente √∫til. Pero resulta que puede hacer *mucho* m√°s. Aqu√≠ hay solo una muestra:

```python
pluto_mass = 1.303 * 10**22
earth_mass = 5.9722 * 10**24
population = 52910390
#         2 puntos decimales 3 puntos decimales, formato como porcentaje separados por comas
"{} pesa alrededor de {:.2} kilogramos ({:.3%} de la masa de la Tierra). Es el hogar de {:,} plutonianos.".format(
    planet, pluto_mass, pluto_mass / earth_mass, population,
)
```

```python
# Refiri√©ndose a los argumentos format() por √≠ndice, comenzando desde 0
s = """Pluto's a {0}.
No, it's a {1}.
{0}!
{1}!""".format('planeta', 'enano planeta')
print(s)
```

# Diccionarios

Los diccionarios son una estructura de datos integrada de Python para asignar claves a valores.

```python
numbers = {'uno':1, 'dos':2, 'tres':3}
```

En este caso `'uno'`, `'dos'` y `'tres'` son las **claves**, y 1, 2 y 3 son sus valores correspondientes.

Se accede a los valores a trav√©s de una sintaxis de corchetes similar a la indexaci√≥n en listas y cadenas.

```python
numbers['uno']
```

Podemos usar la misma sintaxis para agregar otra clave, par de valores

```python
numbers['once'] = 11
numbers
```

O para cambiar el valor asociado con una clave existente

```python
numbers['uno'] = 'Pluton'
numbers
```

Python tiene *comprensiones de diccionario* con una sintaxis similar a las comprensiones de listas que vimos en el tutorial anterior.




```python
planets = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'J√∫piter', 'Saturno', 'Urano', 'Neptuno']
planet_to_initial = {planet: planet[0] for planet in planets}
planet_to_initial
```

El operador `in` nos dice si algo es una clave en el diccionario

```python
'Saturno' in planet_to_initial
```

```python
'Betelgeuse' in planet_to_initial
```

Un bucle for sobre un diccionario har√° un bucle sobre sus claves

```python
for k in numbers:
    print("{} = {}".format(k, numbers[k]))
```

Podemos acceder a una colecci√≥n de todas las claves o todos los valores con `dict.keys()` y `dict.values()`, respectivamente.

```python
# Obtenga todas las iniciales, ord√©nelas alfab√©ticamente y col√≥quelas en una cadena separada por espacios.
' '.join(sorted(planet_to_initial.values()))
```

El muy √∫til m√©todo `dict.items()` nos permite iterar sobre las claves y valores de un diccionario simult√°neamente. (En la jerga de Python, un **elemento** se refiere a un par clave-valor)

```python
for planet, initial in planet_to_initial.items():
    print("{} begins with \"{}\"".format(planet.rjust(10), initial))
```
